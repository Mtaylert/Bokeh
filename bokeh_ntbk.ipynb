```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, CategoricalColorMapper
import numpy as np

color_list = ['firebrick','wheat','blue','darkcyan','aqua','magenta']

size_list = [4,6,8,12,18,24]
alpha_list = [1,1,0.8,0.6,0.4,0.2]

p = figure(plot_width=300, plot_height=300)
p.circle(x=[1,2,3,4,5,6], y=[2,-1,8,5,3,4],
        size=size_list, alpha=alpha_list, color=color_list)
output_notebook()
show(p)
```

```python

p = figure(plot_width=300, plot_height=300)
p.line(x=[1,2,3,4,5,6], y=[2,-1,8,5,3,4])
p.line(x=[1,2,3,4,5,6], y=[-2,-1,5,5,3,1], color='red')

output_notebook()
show(p)
```

```python
#patches

class plotting_the_stuff:
    def __init__(self):
        self.plot_size = figure(plot_width=300, plot_height=300)
    def plot_params(self):
        self.plot_size.patch(x=[1,2,3,4,5,6],y=[2,-1,8,5,3,4], alpha=0.5)
        self.plot_size.patch(x=[1,2,3,4,5,6], y=[1,1,1,1,4,6], alpha=0.5, color='red')
        
    def run_program(self):
        self.plot_params()
        output_notebook()
        show(self.plot_size)
        
        
plotting = plotting_the_stuff()
plotting.run_program()





        
```

```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, CategoricalColorMapper

demo = pd.read_excel('demograph.xlsx')

p_basic = figure(plot_width=500, plot_height=300,
          x_axis_label='UPT',
          y_axis_label='% population')


color_mapper = CategoricalColorMapper(factors=['second city', 'urban', 'suburban','town'],
                                     palette=['red','blue','green','magenta'])

p_basic.circle(demo['UPT'],
              demo['Urbanicity'],
              color={'field':'Urbanicity','transform':color_mapper},
              legend='OK')

output_notebook()
show(p_basic)
```
